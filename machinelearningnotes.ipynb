```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 400)
y = x**2

plt.figure(figsize=(8, 6))
plt.plot(x, y, '-r', label='y=x^2')
plt.title('Graph of y=x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right')
plt.grid()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
# Let's visualize vector addition and scalar multiplication
fig, ax = plt.subplots()
ax.quiver(0, 0, 2, 3, angles='xy', scale_units='xy', scale=1, color="r", label="a (2,3)")
ax.quiver(0, 0, 1, -1, angles='xy', scale_units='xy', scale=1, color="b", label="b (1,-1)")
ax.quiver(0, 0, 3, 2, angles='xy', scale_units='xy', scale=1, color="g", label="a+b (3,2)")
ax.quiver(0, 0, 4, 6, angles='xy', scale_units='xy', scale=1, color="y", label="2a (4,6)")

ax.set_xlim(-1, 5)
ax.set_ylim(-2, 7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Vector Operations: Addition and Scalar Multiplication')
ax.grid(True)
ax.legend()

plt.show()
```

```python
# Define matrices A and B
A = np.array([[1, 3], [5, 7]])
B = np.array([[2, 4], [6, 8]])

# Matrix addition
C = A + B

# Scalar multiplication
D = 2 * A

# Print results
print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Addition of A and B (C):\n", C)
print("Scalar multiplication of A by 2 (D):\n", D)
```

## Example of a System
Consider the system of equations:

3 + 4 = 5

2   = 1

```python
# Let's calculate the solution using Python
A = np.array([[3,  4],
              [2, -1]])
b = np.array([5, 1])

# Computing the inverse of A
A_inv = np.linalg.inv(A)

# Calculating x
x = np.dot(A_inv, b)

print("The solution vector x is:\n", x)
```
